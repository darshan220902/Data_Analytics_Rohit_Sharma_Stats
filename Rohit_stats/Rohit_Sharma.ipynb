# Rohit Data Against Aus

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url='https://stats.espncricinfo.com/ci/engine/player/34102.html?class=11;opposition=2;template=results;type=batting;view=innings'
r=requests.get(url)
print(r)
```

```python
scrp=BeautifulSoup(r.text,'lxml')
table=scrp.find_all('table',class_ ='engineTable')
rows=table[3].find_all('tr')
```

```python
headers=[]
for i in rows[0:1]:
    q=i.text.split('\n')

headers= [value for value in q if value != '']

# Now, 'cleaned_data' contains only non-empty strings
print(headers)
```

```python
rohit_data=pd.DataFrame(columns=headers)
rohit_data['Match_id'] = None
z=rohit_data.columns
print(z)
```

```python
j=[]
for k in rows[1:]:
    s=k.text
    p=s.split('\n')
    cleaned_p = [value for value in p if value != '']
    j.append(cleaned_p)

# Now, 'cleaned_data' contains only non-empty strings
#print(j)
```

```python
rohit_data=pd.DataFrame(j,columns=z)
```

```python
print(rohit_data)
```

```python
rohit_data[['Match_Type', 'Opponents_Team']] = rohit_data['Opposition'].str.split('v', expand=True) # spit columns into 2 columns
```

```python
rohit_data.head()
```

```python
rohit_data_clean=rohit_data.drop('Opposition',axis=1)
```

```python
rohit_data_clean.head()
```

```python
rohit_data_clean.to_csv('Rohit_VS_Aus.csv')
```

```python
ro_vs_aus=pd.read_csv('Rohit_VS_Aus.csv')
ro_vs_aus.head()
```

```python
ro_vs_aus.info()
```
